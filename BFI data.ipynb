```python
df = pd.DataFrame()
# specifying the path to csv files
path = r"C:\Users\milan\OneDrive\Desktop\Projects\BFI"

# csv files in the path
file_list = glob.glob(path + "/*.xlsx")
xls_list=glob.glob(path + "/*.xls")

# list of excel files we want to merge.
# pd.read_excel(file_path) reads the excel
# data into pandas dataframe.
excl_list = []
xlsx_list=[]
for file in file_list:
	excl_list.append(pd.read_excel(file))
for xl in xls_list:
    xlsx_list.append(pd.read_excel(xl))

# create a new dataframe to store the
# merged excel file.
excl_merged = pd.DataFrame()
xls_merged=pd.DataFrame()

for excl_file in excl_list:
	
	# appends the data into the excl_merged
	# dataframe.
	excl_merged = excl_merged.append(excl_file, ignore_index=True)

for xls_file in xlsx_list:
	
	# appends the data into the excl_merged
	# dataframe.
	xls_merged = xls_merged.append(xls_file, ignore_index=True)




```

```python
final_list=[excl_merged,xls_merged]
```

```python
df=pd.concat(final_list)
df.info()
```
